```python
from __future__ import print_function

import math
import sys
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import random

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision.transforms as T

from hp import HP

%matplotlib inline

dev = "cuda:0"
```

```python
hp = HP("HHHPPPHHH")
hp.conformation, hp.coordinates()
hp.show(plt.gca())
hp.energy()
hp.random_move(), hp.conformation
```

```python
hp2 = HP("HHHHHPPPPPPHHHHHHPHPHPHHHHHHPPPPPPHHHHHHPHPHPH")
hp2.show(plt.gca())
```
